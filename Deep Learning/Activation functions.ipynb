```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
y = np.arange(-10,10,0.5)
```

```python
y
```

```python
p = sigmoid(y)
```

```python
plt.plot(p)
```

```python
def tanh(z):
  return (np.exp(z) - np.exp(-z)) /(np.exp(z) + np.exp(-z))
```

```python
a = tanh(y)
```

```python
plt.plot(a)
```

```python
def relu(y):
  return np.where(y >= 0, y, 0)
```

```python
a = relu(y)
```

```python
plt.plot(a)
```

```python
def leaky_relu(y):
  return np.where(y >= 0, y, y*0.2)
```

```python
a = leaky_relu(y)
```

```python
plt.plot(a)
```

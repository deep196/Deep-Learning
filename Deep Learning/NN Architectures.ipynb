## Regression

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(3, activation="tanh", input_dim=2))
model.add(Dense(2, activation="tanh"))
model.add(Dense(1))
```

```python
model.summary()
```

## Binary Classification

```python
model = Sequential()
model.add(Dense(3, activation="tanh", input_dim=2))
model.add(Dense(2, activation="tanh"))
model.add(Dense(1, activation="sigmoid"))
```

```python
model.summary()
```

## Multi-class Classification

```python
# 3 class
model = Sequential()
model.add(Dense(3, activation="tanh", input_dim=2))
model.add(Dense(2, activation="tanh"))
model.add(Dense(3, activation="softmax"))
```

```python
model.summary()
```

## Regularization

### L1 and L2

```python
from tensorflow.keras.regularizers import L1, L2
```

```python
model = Sequential()
model.add(Dense(3, activation="tanh", input_dim=2))
model.add(Dense(3, activation="tanh", kernel_regularizer=L1(0.01)))
model.add(Dense(3, activation="tanh", kernel_regularizer=L2(0.01)))
model.add(Dense(2, activation="tanh"))
model.add(Dense(1))
```

### Dropout

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()
model.add(Dense(8, activation="tanh", input_dim=2))
model.add(Dense(8, activation="tanh"))
model.add(Dropout(0.5))
model.add(Dense(4, activation="tanh"))
model.add(Dense(1))
```
